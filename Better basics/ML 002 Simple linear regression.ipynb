```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = (10,5)
plt.rcParams['figure.dpi'] = (150) 
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/aishwaryamate/Datasets/main/NewspaperData.csv')
```

```python
df
```

```python
df[['daily','sunday']]
```

```python
sns.scatterplot(x=df['daily'],y=df['sunday'])
```

```python
sns.heatmap(df.corr(numeric_only=True),annot=True)
```

```python
#import statsmodels
```

```python
import statsmodels.formula.api as smf
```

```python
model=smf.ols('sunday~daily',data=df).fit()
model
```

```python
#model parameter
model.params
```

```python
#print p value
model.pvalues
```

```python
#r2 value
model.rsquared
```

```python
#adjusted r2 value
model.rsquared_adj
```

# predict for new data

```python
test=pd.DataFrame({
    'daily':[100,152,850,654,759]
})
```

```python
test
```

```python
model.predict(test)
```

# Log Transforamtion

```python
df['log_daily']=np.log(df['daily'])
df['log_sunday']=np.log(df['sunday'])
df
```

```python
plt.subplot(1,2,1)
sns.distplot(df['log_daily'])

plt.subplot(1,2,2)
sns.distplot(df['log_sunday'])

plt.tight_layout()
```

```python
df['log_daily'].skew(),df['log_sunday'].skew()
```

# sqrt transforamtion

```python
df['sqrt_daily']=np.sqrt(df['daily'])
df['sqrt_sunday']=np.sqrt(df['sunday'])
df
```

```python
plt.subplot(1,2,1)
sns.distplot(df['sqrt_daily'])

plt.subplot(1,2,2)
sns.distplot(df['sqrt_sunday'])

plt.tight_layout()
```

```python
df['sqrt_daily'].skew(),df['sqrt_sunday'].skew()
```
