# ANOVA test

```python
from scipy import stats
import pandas as pd
pd.set_option('display.max_rows',None)
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/aishwaryamate/Datasets/main/Iris.csv',
              index_col=0)
df
```

```python
f,pval=stats.f_oneway(df['PetalLengthCm'],df['SepalWidthCm'],df['SepalLengthCm'],df['PetalWidthCm'])
```

```python
alpha=0.05
if pval<alpha:
    print('Reject null hypothesis,one samples is different')
else:
    print('fail to reject null hypothesis,all samples are similar')
```

# chi-square test

```python
from scipy import stats
import pandas as pd
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/aishwaryamate/Datasets/main/chi2.csv',
              index_col=0)
df
```

```python
obs=pd.crosstab(index=df['Athlete'],columns=df['Smoker'])
obs
```

```python
chi2,pval,df,exp=stats.chi2_contingency(obs)
```

```python
pval
```

```python
alpha=0.05
if pval<alpha:
    print('Reject null hypothesis,columns are dependent on eachother')
else:
    print('fail to reject null hypothesis,columns are dependent on each other')
```
