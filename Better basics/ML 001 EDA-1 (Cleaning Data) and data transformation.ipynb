2. If NA values are 5% to 40%,fillna(replace)
3. If NA values are greater than 50% in a column,drop the entrire columns

# outlier detection function(creating function)

```python
def outler_detection(data,colname):
    q1 = data[colname].quantile(0.25)
    q3 = data[colname].quantile(0.75)
    iqr= q3 - q1
    
    lower = q1-(1.5*iqr)
    upper = q3+(1.5*iqr)
    
    return print("lower:",lower,"upper:",upper)
```

```python
outler_detection(df,'Ozone')
```
